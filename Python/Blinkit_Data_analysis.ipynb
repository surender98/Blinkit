```python
import pandas as pd
from plotly import graph_objects as go
import plotly.express as px
```

```python
blinkit_df = pd.read_excel('/content/BlinkIT Grocery Data.xlsx')
blinkit_df.head()
```

```python
blinkit_df.tail()
```

```python
blinkit_df.info()
```

```python
blinkit_df.describe()
```

```python
blinkit_df.describe(include='object')
```

```python
blinkit_df['Item Fat Content'] = blinkit_df['Item Fat Content'].replace(['LF','low fat','reg'],['Low Fat','Low Fat','Regular'])
```

```python
blinkit_df['Item Fat Content'].value_counts()
```

```python
Total_Sales = blinkit_df['Sales'].sum()
print(f'Rs.{Total_Sales.round()}')
```

```python
Average_Sales = blinkit_df['Sales'].mean()
print(f'Rs.{Average_Sales.round()}')
```

```python
No_of_Items = blinkit_df['Item Identifier'].count()
print( 'Number of items ',No_of_Items)
```

```python
Average_ratings = blinkit_df['Rating'].mean()
print(f'{Average_ratings.round()}%')
```

```python
sales_by_outlet_year = blinkit_df.groupby('Outlet Establishment Year')['Sales'].sum().reset_index()
fig = px.line(sales_by_outlet_year, x='Outlet Establishment Year', y='Sales', title='Total Sales by Outlet Establishment Year')
fig.update_yaxes(title_text='Total Sales (Thousands)', tickformat=',.0f', ticksuffix='k')
fig.show()
```

```python
sales_by_outlet_size = blinkit_df.groupby('Outlet Size')['Sales'].sum().reset_index()
fig = px.pie(sales_by_outlet_size, values='Sales', names='Outlet Size',
             title='Total Sales by Outlet Size')
fig.show()
```

```python
sales_by_outlet_location = blinkit_df.groupby('Outlet Location Type')['Sales'].sum().sort_values(ascending=False)
stages = sales_by_outlet_location.index.tolist()
values = sales_by_outlet_location.values.tolist()
fig = go.Figure(go.Funnel(
    y = stages,
    x = values,
    textinfo = "value+percent initial"))
fig.update_layout(title='Sales Funnel by Outlet Location')
fig.show()
```

```python
sales_by_outlet_type = blinkit_df.groupby('Outlet Type').agg(
    Total_Sales=('Sales', 'sum'),
    Number_of_Items=('Item Identifier', 'count'),
    Average_Sales=('Sales', 'mean'),
    Average_Rating=('Rating', 'mean')
).reset_index()
sales_by_outlet_type.sort_values(by='Total_Sales', ascending=False, inplace=True)
sales_by_outlet_type
```

```python
total_sales_by_fat_content = blinkit_df.groupby('Item Fat Content')['Sales'].sum().reset_index()
fig = px.pie(total_sales_by_fat_content, values='Sales', names='Item Fat Content',
             title='Total Sales by Item Fat Content')
fig.show()
```

```python
outlet_location_type_fat_content  = blinkit_df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().reset_index()
fig = px.bar(outlet_location_type_fat_content, x='Outlet Location Type', y='Sales', color='Item Fat Content',
             title='Total Sales by Outlet Location Type and Item Fat Content')
fig.show()
```

```python
total_sales_item_type = blinkit_df.groupby('Item Type')['Sales'].sum().reset_index()
fig = px.bar(total_sales_item_type, y='Item Type', x='Sales', title='Total Sales by Item Type',category_orders= {'Item Type': sorted(total_sales_item_type['Item Type'])})
fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.show()
```
